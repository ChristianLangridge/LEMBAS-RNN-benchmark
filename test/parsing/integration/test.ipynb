```python
import numpy as np

big_data = np.load("path/to/your/massive_file.npz", allow_pickle=True)
tiny_data = {}

for key in big_data.files:
    arr = big_data[key]
    if arr.ndim == 1:
        # Slice 1D arrays (like column names) to 100
        tiny_data[key] = arr[:100]
    elif arr.ndim == 2:
        # Slice 2D arrays to 100 samples AND 100 genes
        tiny_data[key] = arr[:100, :100]
    else:
        tiny_data[key] = arr

np.savez("test/parsing/integration/mock_data.npz", **tiny_data)
```

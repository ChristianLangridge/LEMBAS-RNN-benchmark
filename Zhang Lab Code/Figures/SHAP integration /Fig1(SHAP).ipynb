```python
############### DIAGNOSTICS ###############

import torch
import subprocess

print(f"--- 1. PYTORCH INTERNALS ---")
print(f"PyTorch Version: {torch.__version__}")
print(f"PyTorch compiled for CUDA: {torch.version.cuda}")
print(f"Supported Architectures (Arch List): {torch.cuda.get_arch_list()}")

print(f"\n--- 2. HARDWARE & DRIVER ---")
if torch.cuda.is_available():
    major, minor = torch.cuda.get_device_capability(0)
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"Your GPU Compute Capability: {major}.{minor}")
else:
    print("!! CUDA IS NOT AVAILABLE TO PYTORCH !!")

try:
    smi = subprocess.check_output(["nvidia-smi"]).decode("utf-8").split('\n')
    print(f"System Driver Info: {smi[2]}") # Shows Driver Version & Max supported CUDA
except Exception:
    print("Could not run nvidia-smi. Check if drivers are installed.")
```
